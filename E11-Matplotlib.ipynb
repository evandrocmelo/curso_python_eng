```python
import matplotlib.pyplot as plt
import numpy as np

# Dados simulados
umidade = np.random.uniform(20, 80, 50)  # Umidade do solo (%)
produtividade = 50 + 0.8 * umidade + np.random.normal(0, 10, 50)  # Produtividade (kg/ha)

# Criar gr√°fico
plt.figure(figsize=(8, 5))
plt.scatter(umidade, produtividade, c="green", alpha=0.6, edgecolors="black")
plt.title("Correla√ß√£o entre Umidade do Solo e Produtividade", fontsize=14)
plt.xlabel("Umidade do Solo (%)", fontsize=12)
plt.ylabel("Produtividade (kg/ha)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.3)

# Adicionar linha de tend√™ncia
z = np.polyfit(umidade, produtividade, 1)
p = np.poly1d(z)
plt.plot(umidade, p(umidade), "r--", label=f"Tend√™ncia: y = {z[0]:.2f}x + {z[1]:.2f}")
plt.legend()

plt.show()
```

```python
# Dados simulados de pH em diferentes amostras de solo
pH = np.random.normal(6.5, 0.8, 100)  # M√©dia = 6.5, Desvio padr√£o = 0.8

# Criar histograma
plt.figure(figsize=(8, 5))
plt.hist(pH, bins=15, color="brown", edgecolor="black", alpha=0.7)
plt.title("Distribui√ß√£o de pH em Amostras de Solo", fontsize=14)
plt.xlabel("pH do Solo", fontsize=12)
plt.ylabel("Frequ√™ncia", fontsize=12)
plt.axvline(np.mean(pH), color="red", linestyle="dashed", linewidth=1.5, 
            label=f"M√©dia: {np.mean(pH):.2f}")
plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados simulados - crescimento de milho ao longo de 30 dias
dias = np.arange(0, 30)
altura = 0.5 * dias + np.random.normal(0, 0.5, 30)

# Eventos importantes
eventos = {
    "Aplica√ß√£o NPK": 7,
    "Irriga√ß√£o": 14,
    "Colheita": 28
}

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(dias, altura, 'g-', linewidth=2, label='Crescimento do milho')

# Adicionando anota√ß√µes para cada evento
for evento, dia in eventos.items():
    y_pos = altura[dia] + 0.5  # Ajuste para posicionar o texto
    ax.annotate(evento, 
                xy=(dia, altura[dia]),
                xytext=(dia, y_pos + 2),
                arrowprops=dict(facecolor='black', arrowstyle='->'),
                bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                ha='center')

ax.set_title('Crescimento do Milho com Eventos Agr√≠colas', fontsize=14)
ax.set_xlabel('Dias ap√≥s plantio', fontsize=12)
ax.set_ylabel('Altura (cm)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

# Gerando dados simulados para 90 dias
dias = 90
datas = [datetime(2023, 1, 1) + timedelta(days=i) for i in range(dias)]

# Dados simulados - temperatura e umidade
temp = 20 + 10 * np.sin(np.linspace(0, 3*np.pi, dias)) + np.random.normal(0, 1, dias)
umid = 50 + 20 * np.sin(np.linspace(0, 2*np.pi, dias)) + np.random.normal(0, 3, dias)

# Dividindo em esta√ß√µes
primavera = slice(0, 30)
verao = slice(30, 60)
outono = slice(60, 90)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gr√°fico 1: Compara√ß√£o de temperaturas
ax1.plot(datas[primavera], temp[primavera], 'g-', label='Primavera')
ax1.plot(datas[verao], temp[verao], 'r-', label='Ver√£o')
ax1.plot(datas[outono], temp[outono], 'b-', label='Outono')
ax1.set_title('Varia√ß√£o de Temperatura por Esta√ß√£o', fontsize=14)
ax1.set_ylabel('Temperatura (¬∞C)', fontsize=12)
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.3)

# Gr√°fico 2: Compara√ß√£o de umidade
ax2.plot(datas[primavera], umid[primavera], 'g-', label='Primavera')
ax2.plot(datas[verao], umid[verao], 'r-', label='Ver√£o')
ax2.plot(datas[outono], umid[outono], 'b-', label='Outono')
ax2.set_title('Varia√ß√£o de Umidade por Esta√ß√£o', fontsize=14)
ax2.set_ylabel('Umidade Relativa (%)', fontsize=12)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.3)

# Formatar eixos de data
for ax in [ax1, ax2]:
    ax.xaxis.set_major_formatter(plt.FixedFormatter([d.strftime('%d/%m') for d in datas][::15]))
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
